```python
from IPython.display import display
import ipywidgets as widgets
from PIL import Image as PILImage

def show_image_with_values(image_path):
    # Read the image using PIL
    image = PILImage.open(image_path)
    
    # Display the image
    display(image)
    
    # Create input widgets for user interaction
    x_values = widgets.Text(description='X Values:', value='1,2,3,4')
    y_values = widgets.Text(description='Y Values:', value='5,6,7,8')

    # Define function to update plot based on user input
    def update_plot(change):
        x = [float(val.strip()) for val in x_values.value.split(',')]
        y = [float(val.strip()) for val in y_values.value.split(',')]
        print("X Values:", x)
        print("Y Values:", y)

    x_values.observe(update_plot, names='value')
    y_values.observe(update_plot, names='value')

    # Create input widgets layout
    inputs_layout = widgets.VBox([
        widgets.HBox([x_values, y_values]),
    ])

    # Display the input widgets to the right side of the image
    display(inputs_layout)

# Example usage
show_image_with_values(figures_address / fig_name)
```

```python
import json
from IPython.display import display
import ipywidgets as widgets
from PIL import Image as PILImage

def show_image_with_values(image_path):
    # Read the image using PIL
    image = PILImage.open(image_path)
    
    # Display the image
    display(image)
    
    # Create input widgets for user interaction
    x_values = widgets.Text(description='X Values:', value='1,2,3,4')
    y_values = widgets.Text(description='Y Values:', value='5,6,7,8')

    # Define function to update plot based on user input
    def update_plot(change):
        x = [float(val.strip()) for val in x_values.value.split(',')]
        y = [float(val.strip()) for val in y_values.value.split(',')]
        print("X Values:", x)
        print("Y Values:", y)
        
        # Create dictionary with filename and values
        data = {
            'filename': image_path,
            'x_values': x,
            'y_values': y
        }
        
        # Write dictionary to a JSON file
        with open('image_data.json', 'w') as json_file:
            json.dump(data, json_file)

    x_values.observe(update_plot, names='value')
    y_values.observe(update_plot, names='value')

    # Create input widgets layout
    inputs_layout = widgets.VBox([
        widgets.HBox([x_values, y_values]),
    ])

    # Display the input widgets to the right side of the image
    display(inputs_layout)

# Example usage
show_image_with_values(figures_address / fig_name)
```
